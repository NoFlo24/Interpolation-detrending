```python
#can be applied to all datasets
x=np.array(df['Time'],dtype=np.float64)
y=np.array(df['Amplitude'],dtype=np.float64)
```

```python
#can be applied to all datasets
plt.plot(x, y, c='blue')
plt.xlabel('Time (ms)')
plt.ylabel('Amplitude')
plt.show()
```

```python
y_f=interp1d(x, y, kind='linear')
```

```python
#defaults to 8 coordinates to summarize the entire dataset similarly to originlab
amount_of_nodes=int(input("Enter the amount of nodes: "))
coordinates=np.linspace(np.amin(x), np.amax(x), amount_of_nodes)
#386, 479942
y_coordinates=y_f(coordinates)





#user is able to input x values depending on the amount of nodes they entered
for i in range(len(coordinates)):
    coordinates[i]=int(input("Enter x-coordinate: "))


#user is able to input y values depending on the amount of nodes they entered 
for j in range(len(y_coordinates)):
    y_coordinates[j]=float(input("Enter y-coordinate: "))


y_coord=np.array(y_coordinates)
x_coord=np.array(coordinates)
print(x_coord)
np.sort(x_coord)
print(x_coord)
new_y_coord=np.interp(x, x_coord, y_coord)
print(new_y_coord)
```
